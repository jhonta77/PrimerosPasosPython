```python
def NumeroBinario(numero):
    if numero <= 0:
        return numero
    numeroInt = int(numero)
    numeroDec = numero - numeroInt 
    if numeroInt == 0:
        binarioInt = "0"
    else:
        binarioInt = "" 
        while numeroInt > 0:
            residuo = numeroInt % 2
            binarioInt = str(residuo) + binarioInt
            numeroInt = numeroInt // 2
    if numeroDec == 0:
        return binarioInt
    binarioDec = ""
    while numeroDec > 0 :
        numeroDec *= 2
        if numeroDec >=1 :
            binarioDec += "1"
            numeroDec -= 1
        else:
            binarioDec += "0"

    return binarioInt + "." + binarioDec
    '''
    Esta función recibe como parámetro un número entero mayor ó igual a cero y lo devuelve en su 
    representación binaria. Debe recibir y devolver un valor de tipo entero.
    En caso de que el parámetro no sea de tipo entero y mayor a -1 retorna nulo.
    '''
    
    return None
```
