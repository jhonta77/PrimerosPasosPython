```python
import os
RutaArchivo = r"C:\Users\jhont\OneDrive\Escritorio\modulo 1\Clase 02\Emisiones_CO2.csv"
with open(RutaArchivo, 'r', encoding='ISO-8859-1') as file:
    lineas = file.readlines()
```

```python
dicc_emisiones = {
    'cod_pais': [],
    'nom_pais': [],
    'region': [],
    'anio': [],
    'co2': [],
    'co2_percapita': []
}

for linea in lineas[1:]:
    valores = linea.strip().split('|')
    
    if len(valores) == 6:
        # Procesar valores de CO2 y CO2 per cápita
        co2_value = valores[4].replace('.', '').replace(',', '.') if valores[4] else '0'
        co2_percapita_value = valores[5].strip().replace('.', '').replace(',', '.') if valores[5] else '0'
        
        # Agregar valores al diccionario
        dicc_emisiones['cod_pais'].append(valores[0])
        dicc_emisiones['nom_pais'].append(valores[1])
        dicc_emisiones['region'].append(valores[2])
        dicc_emisiones['anio'].append(valores[3])
        try:
            dicc_emisiones['co2'].append(float(co2_value))
            dicc_emisiones['co2_percapita'].append(float(co2_percapita_value))
        except ValueError:
            dicc_emisiones['co2'].append(0.0)
            dicc_emisiones['co2_percapita'].append(0.0)
```

```python
num_variables = len(dicc_emisiones)
print(f'Número de variables: {num_variables}')
```

```python
tipos_datos = {
    'cod_pais': 'str',
    'nom_pais': 'str',
    'region': 'str',
    'anio': 'int',
    'co2': 'float',
    'co2_percapita': 'float'
}

print('Tipos de datos:', tipos_datos)
```

```python
tipos_variables = {
    'cod_pais': 'Categoría',
    'nom_pais': 'Categoría',
    'region': 'Categoría',
    'anio': 'Cuantitativa discreta',
    'co2': 'Cuantitativa continua',
    'co2_percapita': 'Cuantitativa continua'
}

print('Tipo de variables:', tipos_variables)
```

```python
# Inicializamos las variables para las sumas
total_co2 = 0
total_co2_percapita = 0
registros_encontrados = 0

# Iteramos sobre los índices
for i in range(len(dicc_emisiones['region'])):
    # Verificamos la región y el año
    if (dicc_emisiones['region'][i] == 'América Latina y Caribe' 
            and dicc_emisiones['anio'][i] == '2010'):  # Nota: año como string
        
        registros_encontrados += 1
        
        # Obtenemos los valores de CO2 y CO2 per cápita
        valor_co2 = dicc_emisiones['co2'][i]
        valor_co2_percapita = dicc_emisiones['co2_percapita'][i]
        
        # Verificamos que los valores no sean 0 (valores vacíos)
        if valor_co2 != 0:
            total_co2 += float(valor_co2)
        if valor_co2_percapita != 0:
            total_co2_percapita += float(valor_co2_percapita)
        
        # Imprimimos cada registro encontrado para verificación
        print(f"\nRegistro encontrado:")
        print(f"País: {dicc_emisiones['nom_pais'][i]}")
        print(f"CO2: {valor_co2}")
        print(f"CO2 per cápita: {valor_co2_percapita}")

# Imprimimos los resultados
print(f"\nResultados para América Latina y Caribe en 2010:")
print(f"Total de registros encontrados: {registros_encontrados}")
print(f"Total de emisiones de CO2 (kt): {total_co2:,.2f}")
print(f"Total de CO2 per cápita: {total_co2_percapita:,.2f}")
```

```python
# Calcular el total de emisiones de CO2 para América Latina y Caribe en 2010
total_co2 = sum(dicc_emisiones['co2'][i] for i in range(len(dicc_emisiones['region'])) 
                 if dicc_emisiones['region'][i] == 'América Latina y Caribe' 
                 and dicc_emisiones['anio'][i] == "2010")

# Imprimir el resultado
print(f'Total de emisiones de CO2 para América Latina y Caribe en 2010: {total_co2}')

```
